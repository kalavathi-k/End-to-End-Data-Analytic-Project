```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('/content/Banking.csv')
df.head(5)
```

```python
df['Estimated Income'].min()
```

```python
bins = [0, 100000, 300000, float('inf')]
labels = ['Low', 'Med', 'High']

df['Income Band'] = pd.cut(df['Estimated Income'], bins=bins, labels=labels, right=False)
```

```python
df['Income Band'].value_counts().plot(kind='bar')
```

```python
# Examine the distribution of unique cataegories in categorical columns
categorical_cols = df[["BRId", "GenderId", "IAId", "Amount of Credit Cards", "Nationality", "Occupation", "Fee Structure", "Loyalty Classification", "Properties Owned", "Risk Weighting", "Income Band"]].columns

for col in categorical_cols:
  print(f"value counts for '{col}':")
  display(df[col].value_counts())
```

# Univariate Analysis

```python

for i, predictor in enumerate(df[["BRId", "GenderId", "IAId", "Amount of Credit Cards", "Nationality", "Occupation", "Fee Structure", "Loyalty Classification", "Properties Owned", "Risk Weighting", "Income Band"]].columns):
    plt.figure(i)
    sns.countplot(data=df, x=predictor)
```

# Bivariate Analysis

```python
for i, predictor in enumerate(df[[
    "BRId", "GenderId", "IAId", "Amount of Credit Cards", "Nationality",
    "Occupation", "Fee Structure", "Loyalty Classification",
    "Properties Owned", "Risk Weighting", "Income Band"
]].columns):

    plt.figure(i, figsize=(8,4))
    sns.countplot(data=df, x=predictor, hue="Nationality", palette="Set2")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
# HIstplot of value counts for different Occupation

for col in categorical_cols:
  if col == "Occupation":
    continue
  plt.figure(figsize=(8,4))
  sns.histplot(df[col])
  plt.title('Histogram of Occupation Count')
  plt.xlabel(col)
  plt.ylabel("Count")
  plt.show()
```

# Numarical Analysis

```python
numerical_cols = ['Estimated Income', 'Superannuation Savings', 'Credit Card Balance', 'Bank Loans', 'Bank Deposits', 'Checking Accounts', 'Saving Accounts', 'Foreign Currency Account', 'Business Lending']

# Univariate analysis and visualization
plt.figure(figsize=(15,10))
for i,col in enumerate(numerical_cols):
  plt.subplot(4,3,i+1)
  sns.histplot(df[col],kde=True)
  plt.title(col)
plt.show()
```

# Insight of EDA
Customers with higher income and stronger financial status keep more money across deposits, savings, and checking accounts. Middle-aged customers use more banking services compared to younger customers. Customers with foreign currency accounts usually hold higher overall balances. Those with higher credit card balances also tend to take more bank loans, showing some credit dependency. Low-income customers generally maintain low balances and fall under higher credit-risk categories.
